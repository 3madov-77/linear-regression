```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./datasets_Salary.csv')
df = df[['YearsExperience','Salary']]

df.head(5)
```

```python
describe = df.describe()
info = df.info()

print(describe)
print(info)
```

```python

b = df['YearsExperience'].values.reshape(-1,1)

y = df['Salary'].values
```

```python
b_train, b_test, y_train, y_test = train_test_split( b, y, train_size=0.8, test_size=0.2 ,random_state=100)
```

```python
model = LinearRegression()
model.fit(b_train, y_train)
```

```python
y_predict = model.predict(b_test)
```

```python
model.score(b_test, y_test)*100
```

```python
model.score(b_train, y_train)*100
```

```python
plt.scatter(b_train, y_train, color='blue')
plt.plot(b_test, y_predict, color='red')
plt.ylabel('Salary')
plt.xlabel('Years Of Expreince')
plt.title('Training Data vs Model')
plt.show()
```
